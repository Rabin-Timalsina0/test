```python
import numpy as np
from tensorflow.keras.preprocessing.image import ImageDataGenerator
from tensorflow.keras import layers
import tensorflow as tf
import pandas as pd
import matplotlib.pyplot as plt
from PIL import Image
import os
```

```python
df = pd.read_csv('paths.csv')
```

```python
def get_image(path):
    img = Image.open(path).convert('L')
    img = img.resize((128,128))
    img_arr = np.array(img)
    return img_arr
```

```python
images = []
for image in df['file_path']:
    img = get_image(image)
    images.append(img)
```

```python
X = np.array(images)
```

```python
y = df.label
```

```python
y
```

```python
from tensorflow.keras.preprocessing.image import ImageDataGenerator
import numpy as np

datagen = ImageDataGenerator(
    rotation_range=10,
    width_shift_range=0.1,
    height_shift_range=0.1,
    zoom_range=0.1,
    horizontal_flip=True,
    brightness_range=[0.7, 1.15]
)

X = np.array(X)
y = np.array(y)

X = X.reshape(X.shape[0], 128, 128, 1)

unique, counts = np.unique(y, return_counts=True)
minority_class = unique[np.argmin(counts)]
majority_class = unique[np.argmax(counts)]

minority_count = counts[np.argmin(counts)]
majority_count = counts[np.argmax(counts)]

images_to_generate = majority_count - minority_count

augmented_X = []
augmented_y = []

for i in range(len(X)):
    image, label = X[i], y[i]

    image = np.expand_dims(image, axis=0)

    if label == minority_class and images_to_generate > 0:
        aug_iter = datagen.flow(image, batch_size=1)
        aug_image = next(aug_iter)[0]
        augmented_X.append(aug_image)
        augmented_y.append(label)
        images_to_generate -= 1

    augmented_X.append(image[0])
    augmented_y.append(label)


X_augmented = np.array(augmented_X) / 255
y_augmented = np.array(augmented_y)

print("Original dataset size:", X.shape, y.shape)
print("Augmented dataset size:", X_augmented.shape, y_augmented.shape)
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X_augmented,y_augmented,test_size=0.1)
```

```python
data_augmentation = tf.keras.Sequential([
    layers.RandomRotation(0.15), 
    layers.RandomFlip("horizontal"),
    layers.RandomBrightness(0.2),
    layers.RandomContrast(0.2),
])
```

```python
model = tf.keras.Sequential([
    data_augmentation,
    layers.Conv2D(64, 2, activation='relu', input_shape=(128, 128, 1)),
    layers.MaxPooling2D(),
    layers.Conv2D(32, 2, activation='relu'),
    layers.MaxPooling2D(),
    layers.Flatten(),
    layers.Dense(64, activation='relu'),
    layers.Dense(1, activation='softmax') 
])

model.compile(optimizer='adam', loss='binary_crossentropy', metrics=['accuracy'])

model.fit(X_train, y_train, epochs=25, batch_size=32, validation_data=(X_test, y_test))
```
